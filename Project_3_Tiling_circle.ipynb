```python
import cv2
import numpy as np

def nothing(x):
    pass

# Create a black image, a window
img = np.zeros((300,512,3), np.uint8)
cv2.namedWindow('image')

# create trackbars for color change
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)

# create switch for ON/OFF functionality
switch = '0 : OFF \n1 : ON'
cv2.createTrackbar(switch, 'image',0,1,nothing)

while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

    # get current positions of four trackbars
    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G','image')
    b = cv2.getTrackbarPos('B','image')
    s = cv2.getTrackbarPos(switch,'image')

    if s == 0:
        img[:] = 0
    else:
        img[:] = [b,g,r]

cv2.destroyAllWindows()
```

```python
bgr_img = cv2.imread('Mona-Lisap.jpg')
b,g,r = cv2.split(bgr_img)
rgb_img = cv2.merge([b,g,r]) 
rgb_img.shape
height,width = rgb_img.shape[:2]
img = np.zeros((height,width,3), np.uint8)
img.shape
```

```python
radius = 13
n2 = 125
dia = radius*2
h = int(1.732*radius)
height = height + dia
width = width + dia
count = 0
```

```python
def average(i,j,h,bgr,dia):
    add = 0
    for e in range(j,j+dia):
        for f in range(i,i+dia):
            add = add + rgb_img[e,f,bgr]
    add = (add/(dia*dia)//n2)*n2 + 50
    return add
```

```python
for i in range(1,width -dia*2 ,dia):
    for j in range(1,height -dia*2,h*2):
        b_avg = average(i,j,h,0,dia)
        g_avg = average(i,j,h,1,dia)
        r_avg = average(i,j,h,2,dia)
        cv2.circle(img,(i+radius,j+radius), radius, (b_avg,g_avg,r_avg), -1)
        count = count + 1
```

```python
for i in range(1+radius,width -dia*2 ,dia):
    for j in range(1+ h ,height -dia*2,h*2):
        b_avg = average(i,j,h,0,dia)
        g_avg = average(i,j,h,1,dia)
        r_avg = average(i,j,h,2,dia)
        cv2.circle(img,(i+radius,j+radius), radius, (b_avg,g_avg,r_avg), -1)
        count = count + 1
```

```python
cv2.imshow('rgb_img',img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
cv2.imwrite('butterfly2.png',img)
count
```

```python
import collections
counter=collections.Counter(colors)
print(counter)
```

```python
dic = dict(counter)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(dic, orient='index').reset_index()
```

```python
df
```

```python
blank_image = np.zeros((3000,300,3), np.uint8)
```

```python
 font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
i = 1
for key in dic:
    b_p,g_p,r_p = key
    b_p = int(b_p)
    g_p = int(g_p)
    r_p = int(r_p)
    string = str(str(key) +'  count:' +str(df[0][i-1]))
    cv2.rectangle(blank_image,(1,i*30),(30,i*30+25),(b_p,g_p,r_p),-1)
    cv2.putText(blank_image,string,(40,i*30+20), font, .5,(255,255,255),1,cv2.LINE_AA)
    i = i+ 1
    
```

```python
cv2.imshow('rgb_img',blank_image)
cv2.waitKey(0)
cv2.destroyAllWindows() 
cv2.imwrite('mona-pal2.png',blank_image)
```
